```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# Load dataset
data = pd.read_csv("instagram_accounts.csv")

# Features and target
X = data.drop("fake", axis=1)   # Drop target column
y = data["fake"]                # Target column

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Model
model = GradientBoostingClassifier()
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("instagram_accounts.csv")

# Features and target
X = data.drop("fake", axis=1)   # Drop target column
y = data["fake"]                # Target column (0 = genuine, 1 = fake)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Model
model = GradientBoostingClassifier()
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Evaluation
print("âœ… Accuracy:", accuracy_score(y_test, y_pred))
print("\nðŸ“Š Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nðŸ“„ Classification Report:\n", classification_report(y_test, y_pred))

# ðŸ”¹ Visualize Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Genuine", "Fake"], yticklabels=["Genuine", "Fake"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix Heatmap")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example classification counts
labels = ["Fake", "Spammer", "Genuine"]
counts = [120, 80, 300]

# Pie Chart
plt.pie(counts, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("Instagram Account Types")
plt.show()

# Bar Chart
sns.barplot(x=labels, y=counts)
plt.title("Account Classification Distribution")
plt.show()
```
